## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.


# Step 1: Handle incorrect data types

# the values are 'Poor', 'Standard', and 'Good'. We can map them to an ordered numerical scale.
# This is called Ordinal Encoding.
credit_score_mapping = {'Poor': 0, 'Standard': 1, 'Good': 2}
df['Credit_Score'] = df['Credit_Score'].map(credit_score_mapping)


# Step 2: Drop identifier columns
# These columns are unique to each customer and don't provide predictive value.
# SSN is also sensitive data that should be removed.
df.drop(['ID', 'Customer_ID', 'SSN'], axis=1, inplace=True)

print("\n--- Data Information After Cleaning ---")
df.info()

print("\n--- First 5 Rows of Cleaned Data ---")
print(df.head())
````

#### Chart - 7

````python
# Chart - 7 visualization code
# --- KDE Plot to Compare Distributions ---

plt.figure(figsize=(12, 7))
sns.kdeplot(data=df, x='Monthly_Balance', hue='Credit_Score', 
            fill=True, common_norm=False, palette='viridis', alpha=0.5)
plt.title('Distribution of Monthly Balance by Credit Score', fontsize=15)
plt.xlabel('Monthly Balance')
plt.ylabel('Density')
plt.legend(title='Credit Score', labels=['Good', 'Standard', 'Poor'])
plt.show()
````

##### 1. Why did you pick the specific chart?

A Kernel Density Estimate (KDE) plot is a more advanced version of a histogram. It creates a smooth curve to represent the data's distribution. Using the hue parameter allows us to overlay the distributions for different categories (like Credit_Score) on a single chart, making it excellent for comparing the shape and peaks of distributions between groups.

##### 2. What is/are the insight(s) found from the chart?

We can see how the distribution of Monthly_Balance differs for customers with 'Poor', 'Standard', and 'Good' credit. For instance, customers with 'Good' scores might have a distribution that is tightly centered around a lower monthly balance, while customers with 'Poor' scores might have a flatter, more spread-out distribution, indicating less predictable financial behavior.
